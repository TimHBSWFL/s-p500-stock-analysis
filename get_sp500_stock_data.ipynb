```python
import os
import time
import numpy as np
import pandas as pd
import datetime as dt
import yfinance as yf
import mplfinance as mpf
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
path = r"C:\Users\tokud\Projects\Stocks\S&P 500\SP500_Tickers.csv"

df = pd.read_csv(path, dtype='str',encoding='us-ascii',encoding_errors='ignore')
print(df.shape)
df.head()
```

```python
tickers = df['Symbol'].unique().tolist()
tickers = [t.replace(".", "-") for t in tickers]
```

```python
endDate = dt.datetime.now()
startDate = endDate - dt.timedelta(days = 365*5)
endDate, startDate
```

```python
data = yf.download(
    tickers,
    start=startDate,
    end=endDate,
    auto_adjust=True,
    threads=False
)
```

```python
df_close = data['Close']
df_close
```

```python
output_dir = Path(r"C:\Users\tokud\Projects\Stocks\S&P 500")
output_dir.mkdir(parents=True, exist_ok=True)  

output_path = output_dir / "sp500_close_data.csv"

df_close.to_csv(output_path, index=True)
```
